```python
import numpy as np

# 假设 Y 已加载，形状为 (8000, 5)
Y = np.load('Y_labels.npy')  # 替换为实际路径

# 归一化每个向量
norms = np.linalg.norm(Y, axis=1, keepdims=True)  # 每个向量的范数
Y_normalized = Y / norms  # 将每个向量归一化

# 计算所有向量两两之间的夹角
cos_matrix = np.dot(Y_normalized, Y_normalized.T)
# 提取上三角矩阵（去除对角线和重复部分）
i_upper = np.triu_indices_from(cos_matrix, k=1)  # 上三角索引（不包含对角线）
cosine = cos_matrix[i_upper]


#计算所有向量两两之间的欧几里德距离
distances_matrix = np.linalg.norm(Y[:, np.newaxis] - Y, axis=2)
# 提取上三角矩阵（去除对角线和重复部分）
j_upper = np.triu_indices_from(distances_matrix, k=1) 
distance = distances_matrix[i_upper]
```

```python
# 统计大于 0.5 的值及其位置
threshold = 0.5
filtered_indices = np.where(cosine > threshold)[0]  # 筛选满足条件的索引
filtered_values = cosine[filtered_indices]  # 满足条件的相似度值

# 找到对应的位置（在 similarity_matrix 中的行和列索引）
locations = (i_upper[0][filtered_indices], i_upper[1][filtered_indices])


print("Number of values < -0.5:", len(filtered_values))
print("Locations:", locations)


# min(distance)
```
